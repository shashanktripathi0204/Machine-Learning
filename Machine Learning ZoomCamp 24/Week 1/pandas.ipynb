LOC AND ILOC

loc

    Label-based indexing: You use loc to access rows and columns by their labels (the names of the rows and columns).
    Syntax: df.loc[row_label, column_label]
    Example: If you have a DataFrame with an index of names, you can use df.loc['Alice'] to get the data for Alice.

iloc

    Position-based indexing: You use iloc to access rows and columns by their integer positions (like array indexing).
    Syntax: df.iloc[row_index, column_index]
    Example: If you want the first row and second column, you would use df.iloc[0, 1].
